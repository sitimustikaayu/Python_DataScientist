```python
rating = pd.read_csv("ratings_small.csv")
```

```python
# Menampilkan data
rating.head()
```

```python
# Groupby
rating.groupby('movieId')['rating'].median()
```

```python
# Groupby
rating.groupby('movieId').agg({'userId': pd.Series.nunique, 'rating': ['mean', 'median']})
```

```python
movies_rating = rating.groupby('movieId')['rating'].agg(['mean', 'count', 'median'])
movies_rating
```

```python
# Merger
english_Release_rating = english_Release.merge(movies_rating,
                                               left_on = 'id',
                                               right_on = 'movieId',
                                               how = 'inner')
english_Release_rating
```

```python
# profit = revenue - budget
english_Release_rating['profit'] = english_Release_rating['revenue'] - english_Release_rating['budget']
```

```python
# sort film dengan profit yg tinggi
english_Release_rating[['original_title', 'profit', 'runtime']].sort_values('profit', ascending=False).head(10)
```

```python
english_Release_rating['original_language'] == 'en'
```

```python
english_Release_rating['long_duration_movie'] = (english_Release_rating["runtime"] >= 60*3).astype('int')
```

```python
english_Release_rating[english_Release_rating['long_duration_movie'] == 1]
```

```python
english_Release_rating['title_length'] = english_Release_rating['original_title'].str.len()
```

```python
# title yg kurang dari 10
english_Release_rating[english_Release_rating['title_length'] < 5]
```

```python
english_Release_rating['duration_hour'] = english_Release_rating['runtime']/60
```

```python
english_Release_rating['year_release'] = english_Release_rating['release_date'].dt.year
```

```python
english_Release_rating['fav_movie'] = english_Release_rating['median'] >= 3
```

```python
year_release = english_Release_rating.groupby('year_release').size()
```
